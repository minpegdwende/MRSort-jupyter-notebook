```python
# -*- coding: utf-8 -*-

import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import random
from csv import reader, writer
#import scipy
import time
from itertools import chain, combinations
import importlib
import os





#################
# LOADING DATA #
################

# files containing the raw data (file1: partially anonymous products file2: totally anonymous products)
file1 = 'data_inputs/COMPANY_PRODUCTS_anonymized_v1.csv'
file2 = 'data_inputs/COMPANY_PRODUCTS_anonymized_v2.csv'

raw_data = pd.read_csv(file1, sep=";")
# raw_data2 = pd.read_csv(file2, sep=";")



#################
# DATA CLEANING #
#################

data = raw_data

# cleaning rows that are not identified by COMPANY_ID or SIC2_CODE
data = data[np.isfinite(data.SIC2_CODE)]
data = data[np.isfinite(data.COMPANY_ID)]

# cleaning empty rows and empty colums 
data = data.loc[:, (data.isnull().sum(axis=0) < data.shape[0])]
data = data.loc[data.isnull().sum(axis=1) < data.shape[1]-2,:]

# computing reduction rate of the data
reduction_rate = 1 - data.size/raw_data.size


##########################
# DESCRIPTIVE STATISTICS #
##########################

# list of companies
l_companies = data.index.tolist() 

# list of industries
l_industries = list(np.unique(data[['SIC2_CODE']]))

# list of products
l_products = data.columns[2:].tolist()  

# statistic table with mean,std,quantiles,min,max
# data.describe()

# statistics per product
#data.describe().loc[:,"NM1"]

# number of companies (that have at least once purchased a product) per product
nb_companies_per_product = data.describe().loc["count"][2:]

# sum of products per type
sum_products_per_type = data.sum()[2:]

# average number of products per company among those that have been purchased by a least a company
avg_products_per_type = data.mean()[2:]

# total number of products
nb_total_products = sum_products_per_type.sum()

# simplified matrix companies VS products
products = data[data.columns[2:]]

# count the number of products per company
nb_products_per_company = products.count(1)

# sum of products per company
sum_products_per_company = products.sum(1)

# average number of products per company among those with at least one purchase per company
avg_products_per_company = products.mean(1)

# number of companies per industry
nb_companies_per_industry = data[data.columns[:2]].groupby(['SIC2_CODE']).count()

# number of products per industry per type (= a reduced matrix i.e. agregation of companies in industries)
nb_products_per_industry_per_type = data[data.columns[1:]].groupby(['SIC2_CODE']).count()

# number of products per industry
nb_entries_per_industry = nb_products_per_industry_per_type.sum(1)

# sum of products per industry per type per type
sum_products_per_industry_per_type = data[data.columns[1:]].groupby(['SIC2_CODE']).sum()

# sum of products per industry
sum_products_per_industry = sum_products_per_industry_per_type.sum(1)

# mean of sum of products for a company in each industry per type
avg_sum_products_pindustry_ptype = data[data.columns[1:]].groupby(['SIC2_CODE']).mean()

# mean of number of products for a company in each industry per type
avg_nb_products_pindustry_ptype = pd.DataFrame.transpose(pd.DataFrame.transpose(nb_products_per_industry_per_type) / nb_companies_per_industry.COMPANY_ID.tolist())

# average number of entries (in total) per industry = density of each matrix form by a single industry
avg_nb_entries_per_industry = nb_entries_per_industry/(nb_companies_per_industry.COMPANY_ID*products.shape[1])

# sparsity of the matrix of data
sparsity = 1.0 - nb_entries_per_industry.sum() / products.size

# correlation matrix
correlation_matrix = products.corr()


###############################
# STATISTICS FOR REDUCED DATA #
###############################


# first reduced matrix (express by the number of products per industry and per type)
reduced_matrix1 = nb_products_per_industry_per_type.replace(0,np.nan)

# second reduced matrix (express by the sum of products per industry and per type)
# note that for the remaining of the code, reduced_matrix1 is used instead of reduced_matrix2 ; but there are interchangeable 
reduced_matrix2 = sum_products_per_industry_per_type.replace(0,np.nan)

# number of industries (in which a company has already purchased a product) per product
nb_industries_per_product = reduced_matrix1.count()

# sparsity of the reduced matrix
sparsity_reduced_matrix = reduced_matrix1.isnull().sum(1).sum() / reduced_matrix1.size

# correlation reduced matrix
correlation_reduced_matrix = reduced_matrix1.corr()



# This function outputs correlation between pairs of types of products and the number of values that appears on both simultaneously
# order : is the sorting order of the result (0 for the correlation value and 1 for the number of simultaneous values)
def correlated_pairs(data,correlation_matrix,order=0):
    dico = dict()
    for i in range(len(correlation_matrix)):
        for j in range(i):
            if np.isfinite(correlation_matrix.values[i,j]):
                indi = correlation_matrix.index[i]
                indj = correlation_matrix.index[j]
                nnp = len(data[[indi, indj]].loc[data[[indi, indj]].isnull().sum(axis=1) < 1,:])
                #print(i,j,indi,indj,nnp)
                dico[(indi,indj)] = (correlation_matrix.values[i,j],nnp)
    #print(len(dico))
    order = (int(order))%2
    return sorted(dico.items(), key=lambda kv: abs(kv[1][order]), reverse=True)


# applying the previous function on initial matrix and reduced matrix
correlated_pairs_matrix = correlated_pairs(data,correlation_matrix,1)
correlated_pairs_reduced_matrix = correlated_pairs(reduced_matrix1,correlation_reduced_matrix,1)









##################
# PLOTS, OUTPUTS #
##################


# HISTOGRAMS


# This function plot a histogram with annotations given as arguments
# width : width of the bars, ylog : if y is a log scale, xv: whether if we display all xvalues on x axis
def plot_stats(xvalues,yvalues,xlabel = "",ylabel = "",legend = "",ylog = False,xv = True, save_file = "",xrotation="horizontal", width = 1):
    x = np.arange(len(xvalues))
    
    fig, ax = plt.subplots()
    if ylog:
        ax.bar(x, np.log10(yvalues), width)
        #y = np.arange(len(yvalues))
        ylocs, ylab = plt.yticks() 
        #print([int(10**i) for i in ylocs])
        ax.set_yticklabels([int(10**i) for i in ylocs])
    else :
        ax.bar(x, yvalues, width)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(legend)
    
    
    if xv:
        ax.set_xticks(x)
        ax.set_xticklabels(xvalues,rotation=xrotation)
    #ax.legend()
    #autolabel(rects1)
    fig.tight_layout()
    if save_file:
        plt.savefig(save_file)
    plt.show()
    
   
    
# Distribution of the number of companies as a function of the number of non-zero entries
distribution1 = (np.isfinite(products).sum(1)).value_counts()
#plot_stats(distribution1.index.tolist(),distribution1.tolist(),"Nb of non-zero entries","Nb companies", "The number companies in function of non-zero entries",True,True, "plots/distr_nze.png")

# Distribution of the number of industries as a function of the number of non-zero entries
distribution2 = sorted(dict((np.isfinite(reduced_matrix1).sum(1)).value_counts()).items(), key=lambda kv: kv[0], reverse=False)
#plot_stats([i for i,j in distribution2],[j for i,j in distribution2],"Nb of non-zero entries","Nb companies", "The number industries in function of non-zero entries (reduced matrix)",False,True, "plots/distr_nze_rm.png",75)

# Histogram of the number of industries where their number of companies is comprise into a range 
#distribution3 = sorted(dict((nb_companies_per_industry.COMPANY_ID//10).value_counts()).items(), key=lambda kv: kv[0], reverse=False)
distribution3 = nb_companies_per_industry.COMPANY_ID
#plot_stats(distribution3.index.tolist(),distribution3.tolist(),"Industries (one for each bar)","Nb companies", "The number companies per industry",True,False, "plots/distr_nci.png")


# Distribution of the number of entries per industries
distribution4 = nb_entries_per_industry
#plot_stats(distribution4.index.tolist(),distribution4.tolist(),"Industries (one for each bar)","Nb entries", "The number entries per industry",True,False, "plots/distr_nei.png")

# Distribution of the sum of products per industries
distribution5 = sum_products_per_industry
#plot_stats(distribution5.index.tolist(),distribution5.tolist(),"Industries (one for each bar)","Sum of products", "The sum of products per industry",True,False, "plots/distr_spi.png",0)

# Distribution of the number of products per type of products
distribution6 = sum_products_per_type
#plot_stats(distribution6.index.tolist(),distribution6.tolist(),"Types of products","Sum of products", "The number of products per type",True,True, "plots/distr_nppt.png",90)



# CORRELATION MAPS



# This function compute (build and draw) the correlation matrix for a set of given relation between products (with different methods)
def plt_correlation_matrix(products,method,save_file):
    corr_mat = products.corr(method=method)
    sns_plot = sns.heatmap(corr_mat, cmap='coolwarm')
    fig = sns_plot.get_figure()
    fig.savefig(save_file)
    return corr_mat

# application with the initial matrix : products 
#correlation_matrix = plt_correlation_matrix(products,'kendall','plots/correlation_matrix_kendall.png')
#correlation_matrix = plt_correlation_matrix(products,'pearson','plots/correlation_matrix_pearson.png')
#correlation_matrix = plt_correlation_matrix(products,'spearman','plots/correlation_matrix_spearman.png')


#correlation_reduced_matrix = plt_correlation_matrix(reduced_matrix1,'kendall','plots/correlation_reduced_matrix1_kendall.png')
#correlation_reduced_matrix = plt_correlation_matrix(reduced_matrix1,'pearson','plots/correlation_reduced_matrix1_pearson.png')
#correlation_reduced_matrix = plt_correlation_matrix(reduced_matrix1,'spearman','plots/correlation_reduced_matrix1_spearman.png')
#correlation_reduced_matrix = plt_correlation_matrix(reduced_matrix2,'kendall','plots/correlation_reduced_matrix2_kendall.png')
#correlation_reduced_matrix = plt_correlation_matrix(reduced_matrix2,'pearson','plots/correlation_reduced_matrix2_pearson.png')
#correlation_reduced_matrix = plt_correlation_matrix(reduced_matrix2,'spearman','plots/correlation_reduced_matrix2_spearman.png')



# This function runs successively  correlation matrix for all industries with parameters
def plt_correlation_matrices(data, l_industries, method, save_file_format):
    corr_mat_per_ind = dict()
    for i in l_industries:
        #tmp_products = data[data.SIC2_CODE==i][data.columns[2:]].fillna(0)
        tmp_products = data[data.SIC2_CODE==i][data.columns[2:]]
        tmp_cm = tmp_products.corr()
        nb = tmp_cm[np.isfinite(tmp_cm)].count().sum()
        #print(i,nb)
        tmp_cm[np.isfinite(tmp_cm)].count().sum()
        if nb != 0:
            importlib.reload(matplotlib); importlib.reload(plt); importlib.reload(sns)
            plt_correlation_matrix(tmp_products,method,save_file_format + "_" + str(int(i)) + ".png")
            time.sleep(1)
            corr_mat_per_ind[i] = (correlated_pairs(data[data.SIC2_CODE==i],tmp_cm),nb)
    return corr_mat_per_ind

#correlation_matrices_per_industry = plt_correlation_matrices(data,l_industries,'spearman','plots/correlation_matrices/correlation_matrix_spearman')
#correlation_matrices_per_industry = plt_correlation_matrices(data,l_industries,'pearson','plots/correlation_matrices/correlation_matrix_pearson')
#correlation_matrices_per_industry = plt_correlation_matrices(data,l_industries,'kendall','plots/correlation_matrices/correlation_matrix_kendall')




# This function compute a scatter plot according to 2 types of products
def plot_scatter_pair(matrix, p1, p2, save_file):
    matrix.plot.scatter(p1, p2)
    plt.title("Scatter plot - correlation between the number of products (" + p1 + "," + p2 + ")")
    f = open(save_file, "x")
    plt.savefig(save_file)

# plot_scatter_pair(reduced_matrix1, 'NM7', 'NM6', "plots/scatter_plots_reduced_matrix/scatter_red_mat_NM7_NM6.png")

# This function compute a scatter plot according to all pairs of types of products on the given matrix
def plot_scatter_all_pairs(matrix,save_format_file):
    for comb1,comb2 in combinations(l_products,2):
        if len(matrix[[comb1,comb2]].loc[matrix[[comb1,comb2]].isnull().sum(axis=1) < 1,:]) != 0:
            plot_scatter_pair(matrix, comb1, comb2, save_format_file + "_" + comb1 + "_" + comb2 + ".png")
            #time.sleep(0.1)

# plot_scatter_all_pairs(reduced_matrix1, "plots/scatter_plots_reduced_matrix/scatter_red_mat")
# plot_scatter_all_pairs(data, "plots/scatter_plots/scatter")


# This function scatter plots for each industries, for alls pairs of types of products.
def plot_scatter_all_pairs_per_industry(data,sub_path):
    for i in l_industries:
        os.mkdir(sub_path + "/industry_" + str(int(i)))
        plot_scatter_all_pairs(data[data.SIC2_CODE==i],sub_path + "/industry_" + str(int(i)) + "/")
        time.sleep(0.1)

# plot_scatter_all_pairs_per_industry(data, "plots/scatter_plots")




################################
## PREPARING FOR MCDA LEARNING #
################################


# Loading 2 uniques instances from the reduced matrices : first with reduced_matrix1 and the second, reduced_matrix2
# each row (industry) is an alternative
#reduced_matrix1.replace(0,np.nan).to_csv("data_outputs/industries_nb_entries_81_27.csv")
#reduced_matrix2.replace(0,np.nan).to_csv("data_outputs/industries_nb_products_81_27.csv")



# Select randomly a company among th r*100% most dense companies i, ....n terms of number of entries
# cutting columns that contains fewest non-zero values : ration '1-c)*100
# Load (output) data concernning the choosing industry (list of companies and products) as a csv file
random.seed(1)
def load_companies_from_industry(r = 1, c = 1):
    tmp_data = avg_nb_entries_per_industry
    limit = int(len(tmp_data)*r)
    limit = 1 if (limit<1) else limit
    limit = len(tmp_data) if (limit>len(tmp_data)) else limit
    tmp = tmp_data.sort_values(ascending=False).iloc[range(limit)]
    ch = int(random.randint(0,limit-1))
    #print(ch,limit,tmp.iloc[[ch]])
    industry = int(tmp.keys()[ch])
    industry_data = data[data.SIC2_CODE==industry]
    
    #saving
    cols = industry_data.shape[0] if (int(industry_data.shape[0]*c)>industry_data.shape[0]) else int(industry_data.shape[0]*c)
    cols = 1 if (int(industry_data.shape[0]*c)<0) else int(industry_data.shape[0]*c)
    result = industry_data.loc[:, (industry_data.isnull().sum(axis=0) < cols)]
    result.to_csv("data_outputs/companies_industry_"+str(industry)+".csv")
    return industry,industry_data


#res = load_companies_from_industry(0.1,0.8)


# This function extends the previous one with the possibility to construct mixed instances (companies from potentially many industries).
# same arguments 
def load_companies_from_all(r = 1, c = 1):
    tmp_data = data[data.columns[2:]].count(1)
    limit = int(len(tmp_data)*r)
    limit = 1 if (limit<1) else limit
    limit = len(tmp_data) if (limit>len(tmp_data)) else limit
    ch = int(random.randint(0,limit-1))
    ch = 1 if (ch<1) else ch
    tmp = tmp_data.sort_values(ascending=False).iloc[range(ch)]
    #print(ch,limit,tmp)
    #company = int(tmp.keys()[ch])
    #print(limit,ch,tmp)
    companies_data = data.loc[tmp.keys()]
    cols = companies_data.shape[0] if (int(companies_data.shape[0]*c)>companies_data.shape[0]) else int(companies_data.shape[0]*c)
    cols = 1 if (int(companies_data.shape[0]*c)<0) else int(companies_data.shape[0]*c)
    result = companies_data.loc[:, (companies_data.isnull().sum(axis=0) < cols)]
    result.to_csv("data_outputs/companies_all_"+str(len(companies_data))+".csv")
    return companies_data

#res = load_companies_from_all(0.1,0.5)


# Construct instance with a maximal density
# Variant of load_companies_from_all : the difference is that selection is randomized. 
# r and c the same arguments as in previous functions, nb is for  
def load_companies_from_all2(r = 1, nb = 1, c = 1):
    tmp_data = data[data.columns[2:]].count(1)
    limit = int(len(tmp_data)*r)
    limit = 1 if (limit<1) else limit
    limit = len(tmp_data) if (limit>len(tmp_data)) else limit
    #ch = int(random.randint(0,limit-1))
    #ch = 1 if (ch<1) else ch
    tmp = tmp_data.sort_values(ascending=False).iloc[range(limit)]
    nb_drops = len(tmp)-int(len(tmp)*nb)
    nb_drops = len(tmp)-1 if (nb_drops>len(tmp)-1) else nb_drops
    #print(len(tmp))
    for i in range(nb_drops):
        ch = int(random.randint(0,len(tmp)-1))
        tmp = tmp.drop([tmp.keys()[ch]])
    #print(nb_drops,len(tmp),tmp)
    #print(ch,limit,tmp)
    #company = int(tmp.keys()[ch])
    #print(limit,ch,tmp)
    companies_data = data.loc[tmp.keys()]
    
    # Saving in csv file
    cols = companies_data.shape[0] if (int(companies_data.shape[0]*c)>companies_data.shape[0]) else int(companies_data.shape[0]*c)
    cols = 1 if (int(companies_data.shape[0]*c)<0) else int(companies_data.shape[0]*c)
    result = companies_data.loc[:, (companies_data.isnull().sum(axis=0) < cols)]
    result.to_csv("data_outputs/companies_all2_"+str(len(companies_data))+".csv")
    return companies_data

#res = load_companies_from_all2(0.01,0.9,0.3)



# Ouputs an instance readily for learning of MCDA model (in pymcda format in oso code)
# From a instance where alternatives are companies
def build_MCDA_instance(input_file):
    newfile = "/".join(input_file.split('/')[:-1]+[""]) + "pymcda_" + input_file.split('/')[-1]
    out = writer(open(newfile, "w"), delimiter=" ")
    i=0
    for row in reader(open(input_file, "r"), delimiter=","):
        if i==0:
            nbc = len(row)
            criteria = row[3:]
            out.writerow(["criterion,direction," + ("," * (nbc - 4))])
            for crit in criteria:
                out.writerow([crit + ",1," + ("," * (nbc - 4))])
            out.writerow(["," * (nbc - 2)])
            out.writerow(["category,rank," + ("," * (nbc - 4))])
            out.writerow(["recommend,1," + ("," * (nbc - 4))])
            out.writerow(["notrecommend,2," + ("," * (nbc - 4))])
            out.writerow(["," * (nbc - 2)])
            out.writerow(["pt," + ",".join(criteria) + ",assignment"])
        else:
            tmp=['0' if i=='' else i for i in row[3:]]
            #print(tmp)
            if tmp[-1]=='0':
                out.writerow([row[1] + "," + ",".join(tmp) + ",notrecommend"])
            else:
                out.writerow([row[1] + "," + ",".join(tmp) + ",recommend"])
        i+=1
    #print(i,out)
    #out.close()

#build_MCDA_instance("data_outputs/companies_industry_88.csv")
#build_MCDA_instance("data_outputs/companies_all2_30.csv")
#build_MCDA_instance("data_outputs/companies_all2_57.csv")
#build_MCDA_instance("data_outputs/companies_all_233.csv")
#build_MCDA_instance("data_outputs/companies_industry_10.csv")
#build_MCDA_instance("data_outputs/companies_all_1165.csv")
#build_MCDA_instance("data_outputs/companies_all2_55.csv")


# This a variant of the previous function, where choice is given in whether or not the recommended product is part of the set of criteria.
def build_pymcda_instance2(file, decision_product_in=1):
    newfile = "/".join(file.split('/')[:-1]+[""]) + "pymcda_" + str(decision_product_in) + "_" + file.split('/')[-1]
    out = writer(open(newfile, "w"), delimiter=" ")
    i=0
    for row in reader(open(file, "r"), delimiter=","):
        if i==0:
            nbc = len(row)
            criteria = row[1:] if decision_product_in else row[1:-1]
            tmp_add = 1-decision_product_in
            out.writerow(["criterion,direction," + ("," * (nbc - 3-tmp_add))])
            for crit in criteria:
                out.writerow([crit + ",1," + ("," * (nbc - 3-tmp_add))])
            out.writerow(["," * (nbc-tmp_add)])
            out.writerow(["category,rank," + ("," * (nbc - 3-tmp_add))])
            out.writerow(["recommend,1," + ("," * (nbc - 3-tmp_add))])
            out.writerow(["notrecommend,2," + ("," * (nbc - 3-tmp_add))])
            out.writerow(["," * (nbc-tmp_add)])
            out.writerow(["pt," + ",".join(criteria) + ",assignment"])
        else:
            # big assumption -> replacing missing values with 0
            tmp=['0' if i=='' else i for i in row[1:]]
            #print(tmp)
            if tmp[-1]=='0':
                out.writerow(["pt" + row[0] + "," + ",".join(tmp[:int(-1+len(tmp)+decision_product_in)]) + ",notrecommend"])
            else:
                out.writerow(["pt" + row[0] + "," + ",".join(tmp[:int(-1+len(tmp)+decision_product_in)]) + ",recommend"])
        i+=1
    #print(i,out)
    #out.close()

#Building the unique pymcda instances taking industries as an alternatives (with two cases concerning the recommended product) 
#build_pymcda_instance2("data_outputs/industries_nb_entries_81_27.csv",1)
#build_pymcda_instance2("data_outputs/industries_nb_products_81_27.csv",1)
#build_pymcda_instance2("data_outputs/industries_nb_entries_81_27.csv",0)
#build_pymcda_instance2("data_outputs/industries_nb_products_81_27.csv",0)
```
